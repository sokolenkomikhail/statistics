```python
from math import factorial


def combinations(n, k):
    return int(factorial(n)/(factorial(k)*factorial(n - k)))
```

### 1. Из колоды в 52 карты извлекаются случайным образом 4 карты.

a) Найти вероятность того, что все карты – крести. 

```python
# Вероятность того, что все карты - крести:
# количество комбинаций крести деленное на общее количество комбинаций

# Также можно посчитать следующим образом:
# 13/52(1-ая карта крести) * 12/51(2-ая карта крести) * 11/50(3-я карта крести) * 10/49(4-ая карта крести)

print('Вероятность, что все карты - крести:', combinations(13, 4)/combinations(52, 4))
```

б) Найти вероятность, что среди 4-х карт окажется ХОТЯ БЫ один туз.

```python
# Вероятность неудачного исхода (ни одного туза) равна:  
# количество комбинаций не тузов деленное на общее количество комбинаций

# Также можно посчитать следующим образом:
# 48/52(1-ая карта не туз) * 47/51(2-ая карта не туз) * 46/50(3-я карта не туз) * 45/49(4-ая карта не туз)

# Таким образом вероятность удачного исхода (хотя бы один туз) равна: 1 - 'вероятность неудачного исхода'

print('Вероятность того, что среди карт будет хотя бы один туз:', 1 - combinations(48, 4)/combinations(52, 4))
```

### 2. На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

```python
# Вероятность подобрать код с первой попытки равна:
# единица деленная на количество комбинаций из десяти цифр

print('Вероятность открыть дверь с первой попытки:', 1/combinations(10, 3))
```

### 3. В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

```python
# Вероятность достать 3 окрашенные детали:
# количество комбинаций из 9 окрашенных деталей деленное на общее количество комбинаций

# Также можно посчитать следующим образом:
# 9/15(1-ая деталь окрашена) * 8/14(2-ая деталь окрашена) * 7/13(3-я деталь окрашена)

print('Вероятность того, что все извлеченные детали окрашены:', combinations(9, 3)/combinations(15, 3))
```

### 4. В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

```python
# Вероятность приобрести 2 выигрышных билета:
# количество комбинаций из 2-ух билетов деленное на общее количество комбинаций

# Также можно посчитать следующим образом:
# 2/100(1-ый билет выигрышный) * 1/99(2-ой билет выигрышный)

print('Вероятность того, что 2 приобретенных билета - выигрышные:', combinations(2, 2)/combinations(100, 2))
```
